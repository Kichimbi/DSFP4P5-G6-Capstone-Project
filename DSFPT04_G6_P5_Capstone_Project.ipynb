```python
#Loading required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Load Dataset
pump_prices=pd.read_excel('Epra_Data-1.xlsx')
pump_prices.head()
```

```python
from dateutil import parser

# Renaming Columns
pump_prices.rename(columns={'Price Period': 'Date'}, inplace=True)

# Creating Start Date and End Date Columns
pump_prices['Start_Date'] = pump_prices['Date'].str.split('-').str[0]
pump_prices['End_Date'] = pump_prices['Date'].str.split('-').str[-1]

# Function to parse dates with fallback to original data
def parse_date(date_str):
    try:
        return parser.parse(date_str).strftime('%Y-%m-%d')
    except ValueError:
        return date_str

# Apply the parsing function to the date columns
pump_prices['Start_Date'] = pump_prices['Start_Date'].apply(parse_date)
pump_prices['End_Date'] = pump_prices['End_Date'].apply(parse_date)

# Convert the "Start_Date" column to a datetime format
pump_prices['Start_Date'] = pd.to_datetime(pump_prices['Start_Date'], errors='coerce')
# Convert the "End_Date" column to a datetime format
pump_prices['End_Date'] = pd.to_datetime(pump_prices['End_Date'], errors='coerce')

# Calculate the duration in days between start_date and End_date
pump_prices['Duration(Days)'] = pump_prices['End_Date'] - pump_prices['Start_Date']

# Replace wundanji town name with its correct name Wundanyi
pump_prices['Town'] = pump_prices['Town'].replace('wundanji', 'wundanyi')

# Display the modified dataset
pump_prices.head()
```

```python
#Load Dollar DataSet
dollar = pd.read_excel('Key CBK Indicative Exchange Rates (1).xlsx')
dollar.head()
```

```python
#Renaming Column Date to Start_Date
dollar.rename(columns={'Date':'Start_Date'},inplace=True)
#Convert Start date column into date data type column
dollar['Start_Date']= pd.to_datetime(dollar['Start_Date'])
dollar.head()
```

```python
# Merge the two DataFrames based on the 'Start Date' column
combined_df = pd.merge(pump_prices, dollar, on='Start_Date', how='outer')
```

```python
# Rename Date column to match the End_Date column of pump_prices
dollar.rename(columns={'Start_Date': 'End_Date'}, inplace=True)

# Convert End_Date column into date data type column
dollar['End_Date'] = pd.to_datetime(dollar['End_Date'])
```

```python
# Merge the two DataFrames based on the 'End_Date' column
combined_df1 = pd.merge(pump_prices, dollar, on=['End_Date'], how='inner')

combined_df1.info()
```

```python
#Combine the two dataframes now this caputes the dollar prices that fall either on start date or end date
combined_df2 = pd.concat([combined_df, combined_df1], ignore_index=True)
```

```python
#There duplicate values
combined_df2[combined_df2.duplicated()]
#Drop duplicate values
combined_df2 = combined_df2.drop_duplicates()

#View Dataset minus Duplicates
combined_df2.head()
```

```python
#Drop columns we don't need
combined_df2.drop(['Date','Currency','Buy','Sell'],axis=1,inplace=True)
```

```python
# Rename Mean column to Dollar_Price as per the fuel price at that time
combined_df2.rename(columns={'Mean': 'Dollar_Price'}, inplace=True)

# Convert 'Duration(Days)' column to string and remove 'days' string
combined_df2['Duration(Days)'] = combined_df2['Duration(Days)'].astype(str).str.split('days').str[0]

# Convert 'Duration(Days)' column to numeric type
combined_df2['Duration(Days)'] = pd.to_numeric(combined_df2['Duration(Days)'], errors='coerce')

# Drop missing values
combined_df2.dropna(inplace=True)

# Display the first few rows of the modified DataFrame
combined_df2.head()
```
